# 1. Markdown (MD)

## Sintaxis más básica de Markdown:
- Las líneas de texto se separan siempre con *líneas en blanco*, así:

```
primera línea

segunda línea
```

lo que se muestra así:

primera línea

segunda línea


Si no lo hacen, aparecen pegadas, así:

primera línea
segunda línea


- `#` (`##`, `###`, etc.) se usa para marcar niveles de título. Por ejemplo:
```
# Este es un título de nivel 1
## Este es un título de nivel 2
### Este es un título de nivel 3
#### Este es un título de nivel 4
##### Este es un título de nivel 5
```
será representado así:
# Este es un título de nivel 1
## Este es un título de nivel 2
### Este es un título de nivel 3
#### Este es un título de nivel 4
##### Este es un título de nivel 5


- Listas enumeradas se marcan así:

```
1. primer elemento
2. segundo elemento
3. tercer elemento
```

lo que se mostrará así:

1. primer elemento
2. segundo elemento
3. tercer elemento

***


- Listas no enumeradas (viñetas) se marcan con un guión (`-`) o con un asterisco (`*`):

```
- primer elemento
- segundo elemento
- tercer elemento
```

lo que se mostrará así:

- primer elemento
- segundo elemento
- tercer elemento

***


- Las itálicas (o cursivas) se escriben entre asteriscos simples, y las negritas entre asteriscos dobles: 

```
*texto en itálicas*
**texto en negritas**
```

*texto en itálicas*

**texto en negritas**

***

- Para mostrar texto de código en línea, se encierra entre dos caracter de acento grave (&#96;), así: 
```
La siguiente palabra está formateada como código: `este es un código`
```
que se muestra así:

La siguiente palabra está formateada como código: `este es un código`

- Para líneas completas de código (bloques de código), se encierra todo en tres &#96;&#96;&#96;, así:

![mdcode](https://github.com/nivaca/HD201920/raw/master/mdcode.png)

que se muestra así.

```
Primera línea de código
Segunda línea
Tercera línea
...
```
***


- Para introducir links (vínculos) de internet, simplemente se introducen sin formato. Markdown los identifica como tales. Por ejemplo:
```
https://daringfireball.net/projects/markdown/syntax
```
se verá simplemente así: https://daringfireball.net/projects/markdown/syntax


- Para introducir links de internet con texto distinto del la dirección, se usa la siguiente sintaxis:
```
[texto del link](dirección del link)
```

Por ejemplo:
```
[Página de la Universidad de los Andes](https://www.uniandes.edu.co)
```

se mostraría así:

[Página de la Universidad de los Andes](https://www.uniandes.edu.co)

- Para introducir imágenes, la sintaxis es:

```
![texto oculto opcional](dirección remota de la imagen)
```

Por ejemplo:
```
![John Gruber](https://daringfireball.net/graphics/author/addison-bw.jpg)
```

se mostraría así:

![John Gruber](https://daringfireball.net/graphics/author/addison-bw.jpg)

- También es posible introducir código HTML directamente y Markdown lo mostrará correctamente, por ejemplo:
```
<table style="width:100%">
  <tr>
    <th>Firstname</th>
    <th>Lastname</th> 
    <th>Age</th>
  </tr>
  <tr>
    <td>Jill</td>
    <td>Smith</td> 
    <td>50</td>
  </tr>
  <tr>
    <td>Eve</td>
    <td>Jackson</td> 
    <td>94</td>
  </tr>
</table>
```

se mostrará así:

<table>
  <tr>
    <th>Firstname</th>
    <th>Lastname</th> 
    <th>Age</th>
  </tr>
  <tr>
    <td>Jill</td>
    <td>Smith</td> 
    <td>50</td>
  </tr>
  <tr>
    <td>Eve</td>
    <td>Jackson</td> 
    <td>94</td>
  </tr>
</table>

#### POSDATA

Para buscar caracteres unicode (acentuados, etc.), usamos `\p{L}` en Atom.

Por ejemplo:

- Búsqueda: `([ \p{L}])\-`
- Reemplazo: `$1---`

Esto reemplazaría todos los guiones parentéticos, incluyendo los que están junto a un caracter acentuado: e.g. `proseguí-`.

¿Cómo funciona?

1. Empecemos por los paréntesis cuadrados: `[ \p{L}]`. Esto significa: capture cualquier espacio en blanco o cualquier caracter alfanumérico (incluidos los acentuados). Recuerden que los paréntesis cuadrados significan: "cualquiera de los que están aquí". Por ejemplo: `[aeiou]` captura *uno* cualquiera de los siguientes caracteres: `a`, `e`, `i`, `o` o `u`. Entonces `[ \p{L}]` significa: capture *uno* de cualquiera de estos: ` ` o `\p{L}`.
2. Eso está encerrado entre paréntesis: `([ \p{L}])`, lo que significa que eso es un grupo de búsqueda. Como solo hay uno, lo referimos como `$1`.
3. A eso lo sigue `\-`, que es un guión (recuerden que, como el signo `-` tiene un significado en Regex, a saber, un rango, si lo queremos usar literalmente, debemos "escaparlo" con un backslash, `\-`).
4. Entonces, todo esto: `([ \p{L}])\-` significa: busque cualquier guión que esté precedido de un espacio o cualquier caracter alfanumérico (incluidos los acentuados), y guarde lo que lo precede en el grupo `$1`. Esto es importante porque, si no lo guardáramos, se perdería en el reemplazo.
5. El reemplazo, a saber `$1---`, reemplaza el texto buscado con lo que guardó en el grupo (un espacio en blanco o un caracter alfanumérico) y la línea (los tres guiones seguidos: `---`).

¿Qué pasaría no usaran los paréntesis, es decir, el grupo de captura? Pruébenlo ustedes mismos.
