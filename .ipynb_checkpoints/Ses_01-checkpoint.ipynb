# 1. Markdown (MD)

## Sintaxis más básica de Markdown:
- Las líneas de texto se separan siempre con *líneas en blanco*, así:

```
primera línea

segunda línea
```

lo que se muestra así:

primera línea

segunda línea


Si no lo hacen, aparecen pegadas, así:

primera línea
segunda línea


- `#` (`##`, `###`, etc.) se usa para marcar niveles de título. Por ejemplo:
```
# Este es un título de nivel 1
## Este es un título de nivel 2
### Este es un título de nivel 3
#### Este es un título de nivel 4
##### Este es un título de nivel 5
```
será representado así:
# Este es un título de nivel 1
## Este es un título de nivel 2
### Este es un título de nivel 3
#### Este es un título de nivel 4
##### Este es un título de nivel 5


- Listas enumeradas se marcan así:

```
1. primer elemento
2. segundo elemento
3. tercer elemento
```

lo que se mostrará así:

1. primer elemento
2. segundo elemento
3. tercer elemento

***


- Listas no enumeradas (viñetas) se marcan con un guión (`-`) o con un asterisco (`*`):

```
- primer elemento
- segundo elemento
- tercer elemento
```

lo que se mostrará así:

- primer elemento
- segundo elemento
- tercer elemento

***


- Las itálicas (o cursivas) se escriben entre asteriscos simples, y las negritas entre asteriscos dobles: 

```
*texto en itálicas*
**texto en negritas**
```

*texto en itálicas*

**texto en negritas**

***

- Para mostrar texto de código en línea, se encierra entre dos caracter de acento grave (&#96;), así: 
```
La siguiente palabra está formateada como código: `este es un código`
```
que se muestra así:

La siguiente palabra está formateada como código: `este es un código`

- Para líneas completas de código, se encierra todo en tres &#96;&#96;&#96;, así:

![mdcode](https://github.com/nivaca/HD201920/raw/master/mdcode.png)

que se muestra así.

```
Primera línea de código
Segunda línea
Tercera línea
...
```
***


- Para introducir links (vínculos) de internet, simplemente se introducen sin formato. Markdown los identifica como tales. Por ejemplo:
```
https://daringfireball.net/projects/markdown/syntax
```
se verá simplemente así: https://daringfireball.net/projects/markdown/syntax


- Para introducir links de internet con texto distinto del la dirección, se usa la siguiente sintaxis:
```
[texto del link](dirección del link)
```

Por ejemplo:
```
[Página de la Universidad de los Andes](https://www.uniandes.edu.co)
```

se mostraría así:

[Página de la Universidad de los Andes](https://www.uniandes.edu.co)

- Para introducir imágenes, la sintaxis es:

```
![texto oculto opcional](dirección remota de la imagen)
```

Por ejemplo:
```
![John Gruber](https://daringfireball.net/graphics/author/addison-bw.jpg)
```

se mostraría así:

![John Gruber](https://daringfireball.net/graphics/author/addison-bw.jpg)

- También es posible introducir código HTML directamente y Markdown lo mostrará correctamente, por ejemplo:
```
<table style="width:100%">
  <tr>
    <th>Firstname</th>
    <th>Lastname</th> 
    <th>Age</th>
  </tr>
  <tr>
    <td>Jill</td>
    <td>Smith</td> 
    <td>50</td>
  </tr>
  <tr>
    <td>Eve</td>
    <td>Jackson</td> 
    <td>94</td>
  </tr>
</table>
```

se mostrará así:

<table>
  <tr>
    <th>Firstname</th>
    <th>Lastname</th> 
    <th>Age</th>
  </tr>
  <tr>
    <td>Jill</td>
    <td>Smith</td> 
    <td>50</td>
  </tr>
  <tr>
    <td>Eve</td>
    <td>Jackson</td> 
    <td>94</td>
  </tr>
</table>

# 2. Pandoc

Pandoc nos permite convertir una gran cantidad de formatos de texto entre sí. Vean aquí: https://pandoc.org

- Instalen Pandoc en su computador
- Conviértanlo a formato HTML así:
> pandoc coronavirus.md -o coronavirus.html

- O a formato Word así:
> pandoc coronavirus.md -o coronavirus.html

- Si tienen LaTeX instalado (o con la extensión de Atom), pueden convertirlo directamente a PDF:
> pandoc coronavirus.md -o coronavirus.pdf

- Para buscar y reemplazar usando RegEx en Atom (`Cmd+f`), debemos seleccionar la opción apropiada en la ventana de búsqueda:
![atom search](https://github.com/nivaca/HD201920/raw/master/atomfind.png)

***

1. Escriban lo siguiente en el campo de búsqueda: 
```    
c.+s
```

2. Escriban lo siguiente en el campo de búsqueda: 
```    
c\w+s
```

3. En el campo de búsqueda escriban:
```
m\w+ de c\w+s
```

4. En el campo de búsqueda escriban:
```
(m\w+) de (c\w+s)
```

y en el de reemplazo escriban:
```
$2 de $1
```

Hagan click en `Reemplazar Todo (Replace All)`.
